# This World Happiness data is a survey about the state of global happiness. The report ranks more than 150 countries by their happiness levels, and has been published almost every year since 2012. We'll use data collected in the years 2015, 2016, and 2017

Some of the sample questions are :-

•	Which are the happiest and least happy countries and regions in the world?

•	Is happiness affected by region?

•	Did the happiness score change significantly over the past three years?

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
df = pd.read_csv(r"C:\Users\gc531\Downloads\World_Happiness_data\World_Happiness_data\worldhappiness.csv")
df.head()
```

```python
sns.set_style("darkgrid")
plt.rcParams["font.size"] = 15
plt.rcParams["figure.figsize"] = (10,7)
plt.rcParams["figure.facecolor"] = "#FFE5B4"
```

```python
df.info()
```

```python
df.head()
```

```python
data = df.rename({"Unnamed: 0":"Index","Happiness Rank":"Happiness_rank","Happiness Score":"Happiness_score","Economy (GDP per Capita)":"GDP_per_capita","Health (Life Expectancy)":"Life_expectancy","Trust (Government Corruption)":"Corruption","Dystopia Residual":"Dystopia_residual"},axis=1)
```

```python
data.head()
```

```python
d = data.dropna(axis=0)
```

```python
d.head()
```

```python
d.Happiness_rank.describe()
```

```python
d.Happiness_score.describe()
```

# Happiness Scores of different Regions

```python
d.groupby("Region")[["Happiness_score"]].plot.bar()
```

# Plot b/w Happiness Score and GDP per Capita

```python
plt.rcParams["figure.figsize"] = (15,7)
plt.title("Plot b/w Happiness Score and GDP per Capita")
sns.scatterplot(x = d.Happiness_score, y = d.GDP_per_capita, hue = d.Region , s = 200)
plt.legend(loc = "upper left",fontsize = "10")
plt.xlabel("Happiness Score")
plt.ylabel("GDP")
```

# GDP by Region

```python
a = d.groupby("Region")["GDP_per_capita"].sum()
a.plot.bar()
```

```python
a.plot.pie(autopct = "%1.1f%%")
plt.title("GDP by Region")
plt.ylabel(" ")
```

# Total Countries in all regions

```python
b = d.groupby("Region")["Country"].count()
b.plot.bar()
```

```python
b.plot.pie(autopct = "%1.1f%%")
plt.title("Total Countries in all Region")
plt.ylabel(" ")
```

# Perception of Corruptiopn in various Regions

```python
c = d.groupby("Region")[["Corruption"]].mean()
c.plot.bar()
```

```python
plt.rcParams["figure.figsize"] = (12,8)
plt.title("Perception of Corruptiopn in various Regions")
plt.xlabel("Regions", fontsize = 15)
plt.ylabel("Corruption Index", fontsize = 15)
plt.xticks(rotation = 30, ha = "right")
plt.bar(c.index,c.Corruption)
```

# Top 10 and Bottom 10 happiest countries Life Expentancy

```python
e = d.sort_values(by = "Happiness_score").drop_duplicates("Country")
```

```python
top_10 = e.tail(10)
top_10.drop_duplicates("Country")[["Life_expectancy"]].plot.bar()
```

```python
bottom_10 = e.head(10)
bottom_10.drop_duplicates("Country")[["Life_expectancy"]].plot.bar()
```

```python
fig, axes = plt.subplots(1,2, figsize = (16,6))
plt.tight_layout(pad = 2)
xlabels = top_10.Country
axes[0].set_title("Top 10 Happiest Countries Life Expectancy")
axes[0].set_xticklabels(xlabels, rotation = 45, ha = "right")
sns.barplot(x = top_10.Country, y = top_10.Life_expectancy, ax = axes[0])
axes[0].set_xlabel("Country Name")
axes[0].set_ylabel("Life Expectancy")



xlabels = bottom_10.Country
axes[1].set_title("Bottom 10 Happiest Countries Life Expectancy")
axes[1].set_xticklabels(xlabels, rotation = 45, ha = "right")
sns.barplot(x = bottom_10.Country, y = bottom_10.Life_expectancy, ax = axes[1])
axes[1].set_xlabel("Country Name")
axes[1].set_ylabel("Life Expectancy")
```

# Countries with most Perception of Corruption

```python
country = d.sort_values(by = "Corruption").drop_duplicates("Country").tail(10)
country.drop_duplicates("Country")[["Corruption"]].plot.bar()
```

```python
plt.rcParams["figure.figsize"] = (12,8)
plt.title("Countries with most Perception of Corruption")
plt.xlabel("Countries", fontsize = 13)
plt.ylabel("Corruption", fontsize = 13)
plt.xticks(rotation = 30, ha = "right")
plt.bar(country.Country,country.Corruption)
```
